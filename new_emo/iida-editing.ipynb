```python
def emo_compare(new, prev):
    return max([
        new.sad - prev.sad, 
        new.joy - prev.joy,
        new.love - prev.love, 
        new.anger - prev.anger,
        new.fear - prev.fear,
        new.surprise - prev.surprise]
    )
```

```python
class Emo:
    def __init__(self,value):
        self.sad = value[0].get("score")
        self.joy = value[1].get("score")
        self.love = value[2].get("score")
        self.anger = value[3].get("score")
        self.fear = value[4].get("score")
        self.surprise = value[5].get("score")
```
